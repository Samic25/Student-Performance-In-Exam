```python
#import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read Data into Dataframe
df = pd.read_csv('StudentsPerformance.csv')
```

```python
print("Shape of the Dataset : ",df.shape)
```

Dataset having 1000 rows with 8 feature's

```python
df.head()
```

```python
df.info()
```

5 are the categorical feature and 3 are the numerical feature

```python
catrgorical_feature = [feature for feature in df.columns if df[feature].dtype == 'object']
```

```python
numerical_feature = [feature for feature in df.columns if df[feature].dtype != 'object']
```

# Univariate Analysis

## Numerical Feature Analysis

```python
for feature in numerical_feature:
    sns.histplot(df[feature],stat='frequency')
    plt.show()
```

All numerical features are having normal distribution.

```python
df[numerical_feature].describe()
```

```python
for feature in numerical_feature:
    sns.boxplot(df[feature])
    plt.show()
```

All 3 numerical feature having outliar below Q1.

## Categorical Feature Analysis

```python
for feature in catrgorical_feature:
    sns.countplot(df[feature])
    plt.xticks(rotation=90)
    plt.show()
```

* Female student are more than male students.
* Group C has more students followed by group D while Group A has least student
* only few student gardians have master's degreee .. More student gardians has completed their education up to graduation.
* More number of students eat standard meal.
* more number of students have not complete their test preapration course

# Bi-variate Analysis

## Numerical Feature V/S Numerical Feature Analysis

```python
sns.heatmap(df[numerical_feature].corr(),annot=True,vmax=1,vmin=-1)
```

All three numerical features are highly co-related with each other

```python
sns.pairplot(data=df)
```

## Categorical V/S Numerical Feature Analysis

```python
f, axes = plt.subplots(1, 3,figsize=(20,10))

sns.boxplot(x=df['test preparation course'],y=df['math score'],ax=axes[0],orient='v')


sns.boxplot(x=df['test preparation course'],y=df['reading score'],ax=axes[1],orient='v')


sns.boxplot(x=df['test preparation course'],y=df['writing score'],ax=axes[2],orient='v')
```

```python
f, axes = plt.subplots(3, 1,figsize=(10,15))

sns.boxplot(x=df['parental level of education'],y=df['math score'],ax=axes[0],orient='v')


sns.boxplot(x=df['parental level of education'],y=df['reading score'],ax=axes[1],orient='v')


sns.boxplot(x=df['parental level of education'],y=df['writing score'],ax=axes[2],orient='v')
```

```python
f, axes = plt.subplots(3, 1,figsize=(10,15))

sns.boxplot(x=df['race/ethnicity'],y=df['math score'],ax=axes[0],orient='v')


sns.boxplot(x=df['race/ethnicity'],y=df['reading score'],ax=axes[1],orient='v')


sns.boxplot(x=df['race/ethnicity'],y=df['writing score'],ax=axes[2],orient='v')
```
